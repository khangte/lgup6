```python
try:
    import matplotlib.pyplot as plt
except:
    !pip install matplotlib
    import matplotlib.pyplot as plt

# 사람 형태를 구성하는 좌표 (스틱 피규어)
# 몸통
torso_x = [0, 0]
torso_y = [0, 4]

# 머리
head_center = (0, 4.7)
head_radius = 0.7

# 팔 (양팔 벌린 상태)
left_arm_x = [0, -2]
left_arm_y = [3.5, 2]
right_arm_x = [0, 2]
right_arm_y = [3.5, 2]

# 다리 (한쪽 앞으로, 한쪽 뒤로)
left_leg_x = [0, -1.5]
left_leg_y = [0, -2]
right_leg_x = [0, 1]
right_leg_y = [0, -1.5]

# 그래프 생성
fig, ax = plt.subplots(figsize=(6, 8))

# 몸통
ax.plot(torso_x, torso_y, 'k-', linewidth=3)

# 머리
head = plt.Circle(head_center, head_radius, fill=False, color='black', linewidth=2)
ax.add_patch(head)

# 팔
ax.plot(left_arm_x, left_arm_y, 'k-', linewidth=3)
ax.plot(right_arm_x, right_arm_y, 'k-', linewidth=3)

# 다리
ax.plot(left_leg_x, left_leg_y, 'k-', linewidth=3)
ax.plot(right_leg_x, right_leg_y, 'k-', linewidth=3)

# 축 설정
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```
